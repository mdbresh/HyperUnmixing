```python
import numpy as np

from baseline import baseline

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.decomposition import NMF
```

First, let's load the subsample of the image we'll run NMF on.

```python
im = np.load('subsample_image_64x64.npy')
```

```python
im.shape
```

Let's check that it looks okay.

```python
plt.imshow(im[:,:,1128-750])
plt.colorbar()
plt.show()
```

NMF extracts the non-negative principle component spectra from the hyperspectral image. You can see that the sample image is 64 x 64 x 1128. The spectra at each pixel is 1128 wavenumbers long. 

We have to initialize our model. The parameters are as follows:
- n_components = 2; we know there are 2 component spectra (from 2 component polymers) in this hyperspectral image
- solver = 'cd'; this is the coordinate descent
- tol = 1e-5; this is the tolerance/convergence factor
- max_iter = 1e6; this is how many times the algorithm will iterate to converge to the tolerance
- l1_ratio = 0.2; this is the distribution of the L1 we'll use to take steps down the gradient
- alpha = 1e-3; this is the step size down the gradient
- random_state = 0; this is how we step down the gradient
- verbose = True/False; depending on if you want to see the convergence

```python
nmf = NMF(n_components = 2, solver = 'cd', tol = 1e-5, max_iter=int(1e6), l1_ratio = 0.2, alpha = 1e-3, random_state = 0, verbose = True)
```

```python
im2d = np.zeros((im.shape[0]**2, im.shape[2]))
for i in range(im.shape[2]):
    im2d[:,i] = im[:,:,i].flatten()
```

```python
nmf.fit_transform(X = im2d + np.abs(np.min(im2d)))
```

```python
np.save('raw_comps', nmf.components_)
```

```python
# components = nmf.components_
components = np.load('raw_comps.npy')
```

We have to correct the baseline for our two components. (Each 1 x 1128 vector in the nmf.components_ is our spectra.)

```python
for i in range(components.shape[0]):
    base_i = baseline(components[i,:])
    components[i,:] = components[i,:] - base_i
```

Loading our ground truths for comparison ...

```python
ps_truth = np.load('ps_ground_truth.npy')
dppdtt_truth = np.load('dppdtt_ground_truth.npy')
```

```python
fig, ax = plt.subplots(ncols = 2, figsize = (16,4))
x = np.arange(750, 1878)
ax[0].plot(x, components[0,:], label = 'DPPDTT NMF', color = 'blue', linewidth = 3)
ax[0].plot(x, dppdtt_truth, 'c+', alpha = 0.5, label = 'DPPDTT truth')
ax[0].legend()

ax[1].plot(x, components[1,:], label = 'PS NMF', color = 'red', linewidth = 3)
ax[1].plot(x, ps_truth, 'm+', alpha = 0.5, label = 'PS truth')
ax[1].legend()

plt.show()
```

If you are fitting data that is not looking like your ground truth, try the following:
- decrease your tolerance `tol` factor by an order of magnitude; this will force the algorithm to try to converge to a more accurate value
- decrease your `alpha` stepsize value so your algorithm does not get trapped ping-ponging around without reaching a minimum
